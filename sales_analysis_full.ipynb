```python

import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/Andrei-analyst/sales-data-analysis/main/sales_data_project2.csv'
df = pd.read_csv(url)

df['order_date'] = pd.to_datetime(df['order_date'])

# Daily Revenue
daily_sales = df.groupby('order_date')['total'].sum()
plt.figure(figsize=(12, 5))
daily_sales.plot(kind='line')
plt.title('Daily Revenue')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()

# Monthly Revenue
df['month'] = df['order_date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('month')['total'].sum()
plt.figure(figsize=(10, 4))
monthly_sales.plot(kind='bar', title='Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Revenue by Category 
category_sales = df.groupby('category')['total'].sum()
plt.figure(figsize=(8, 4))
category_sales.plot(kind='bar', title='Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()

# Top 5 Customers by Revenue
top_customers = df.groupby('customer_id')['total'].sum().nlargest(5)
plt.figure(figsize=(8, 4))
top_customers.plot(kind='bar', title='Top 5 Customers by Revenue')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```
